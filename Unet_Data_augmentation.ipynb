## Read all the unet images and apply transform and save it back to folder

[Inpired from here](https://www.geeksforgeeks.org/apply-changes-to-all-the-images-in-given-folder-using-python-pil/)

```python
from PIL import Image 
from PIL import ImageFilter 
import os
```

```python
# path of the folder containing the raw training images 
trainInPath = "/Users/sumeetrohilla/Downloads/unet-master/data/membrane/train/image"
# path of the folder that will contain the modified training images
trainOutPath ="/Users/sumeetrohilla/Desktop/UNet/unet-master/data/membrane/train/image"

# labelled input and outpath (are basically same)
labelInPath   = "/Users/sumeetrohilla/Downloads/unet-master/data/membrane/train/label"
labelOutPath  = "/Users/sumeetrohilla/Desktop/UNet/unet-master/data/membrane/train/label"

def aug_images(): 
#------------------------------------------------

    # apply the rotation to training images
    for imagePath in os.listdir(trainInPath):
        if imagePath.endswith(".png"):
            # imagePath contains name of the image  
            inputPath = os.path.join(trainInPath, imagePath) 

             # inputPath contains the full directory name 
            img = Image.open(inputPath) 

            fullOutPath = os.path.join(trainOutPath, 'rotated270_'+imagePath) 
    #         fullOutPath contains the path of the output 
    #         image that needs to be generated 
            img.rotate(270).save(fullOutPath) 
#------------------------------------------------
    # apply the rotation to labelled images
    for imagePath in os.listdir(labelInPath):
        if imagePath.endswith(".png"):
            # imagePath contains name of the image  
            inputPath = os.path.join(labelInPath, imagePath) 
             # inputPath contains the full directory name 
            img = Image.open(inputPath) 

            fullOutPath = os.path.join(labelOutPath, 'rotated270_'+imagePath) 
    #         fullOutPath contains the path of the output 
    #         image that needs to be generated 
            img.rotate(270).save(fullOutPath)
```

```python
%time
aug_images()
```
